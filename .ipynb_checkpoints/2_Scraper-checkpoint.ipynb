```python
# 网页下载 （捕获异常、服务器错误500重试下载、用户代理）
import urllib
def download(url, user_agent='GoodCrawler', num_retries=2):
    print ('Downloading:',url)
    # 创建请求而不是url， url读取会被禁止, 设置代理(默认代理会被拒)
    req = urllib.request.Request(url, headers={'User-agent':user_agent}) 
    try:
        # 读取后编码成字符串，在后面re.findall中才能匹配
        html = urllib.request.urlopen(req).read().decode('utf-8')
    except urllib.error.HTTPError as e:
        print ('Download error:',e.reason)
        html = None
        if(num_retries > 0):
            # 处理500-600的服务器错误
            if hasattr(e, 'code') and 500 <= e.code <600:
                return download(url, num_retries - 1)
            
    return html

#download('http://httpstat.us/500')

```

```python
# 使用BeautifulSoup获取example页面中的国家面积
from bs4 import BeautifulSoup
def BeautifulSoupScrapingArea():
    url = 'http://example.webscraping.com/places/default/view/Armenia-12'
    html = download(url)
    soup = BeautifulSoup(html)
    tr = soup.find(attrs={'id':'places_area__row'})
    td = tr.find(attrs={'class':'w2p_fw'})
    area = td.text
    print(area)
BeautifulSoupScrapingArea()
```

```python
# 使用lxml模块抓取
import lxml.html

def LxmlScrapingArea():
    url = 'http://example.webscraping.com/places/default/view/Armenia-12'
    html = download(url)
    tree = lxml.html.fromstring(html)
    # 实际实现为调用cssselect，并转换为XPath选择器
    td = tree.cssselect('tr#places_area__row > td.w2p_fw')[0]
    area = td.text_content()
    print(area)
    
    
LxmlScrapingArea()
```
